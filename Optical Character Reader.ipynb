```python
#import the needed packages

import cv2
import pytesseract
from PIL import Image
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pytesseract import Output
```

```python
# read the image with text

img=cv2.imread("D:\\Python\\Text Recognition\\texts.jfif")

#show image

imgplot = plt.imshow(img)
```

```python
#convert to grayscale image

gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#show image

imgplot = plt.imshow(gray)
```

```python
# converting thresh 

Thresh_img = cv2.threshold(gray, 0,255,cv2.THRESH_BINARY| cv2.THRESH_OTSU)[1]

# applying blur
img1 = cv2.medianBlur(gray, 3)

#show image

imgplot = plt.imshow(Thresh_img)
```

```python
d = pytesseract.image_to_data(img1, output_type=Output.DICT)
print(d.keys())

n_boxes = len(d['text'])
for i in range(n_boxes):
    if int(d['conf'][i]) > 60:
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        image = cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), 2)

#show image

imgplot = plt.imshow(image)
```

```python
#memory usage with image i.e. adding image to memory

filename = "{}.jpg".format(os.getpid())
cv2.imwrite(filename, image)

```

```python
text = pytesseract.image_to_string(Image.open(filename))
os.remove(filename)
print(text)
# here we are converting image text to string and selecting the tesseract language as English
# OUTPUT OF OCR IS SHOWN BELOW
```
